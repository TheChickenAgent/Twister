# 20 and 50 shots class weighting techniques

## Weighted results

```python
## 20 shots

import numpy as np
import matplotlib.pyplot as plt

data = [[0.3592, 0.3697, 0.9718], [0.1647, 0.4768, 0.9666], [0.1459, 0.3954, 0.9679], [0.1647, 0.4768, 0.9666]]
X = np.arange(3) # the x locations for the groups
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color="lightcoral", width=0.15)
ax.bar(X + 0.15, data[1], color="wheat", width=0.15)
ax.bar(X + 0.30, data[2], color="sienna", width=0.15)
ax.bar(X + 0.45, data[3], color="darkorange", width=0.15)  

ax.set_xticks(X+0.225, ("ICW", "SRICW", "None"))
ax.legend(labels=["Precision", "Recall", "F1", "Accuracy"])
plt.title("Weighted results class weighting techniques 20-shots")

plt.show()
```
