# Summary techniques

```python
import numpy as np
import matplotlib.pyplot as plt

data = [
    [0.0945, 0.3219, 0.2176, 0.0520, 0.0917],
    [0.1365, 0.4280, 0.4082, 0.1000, 0.1079],
    [0.5303, 0.6575, 0.7182, 0.6920, 0.6421],
    [0.7422, 0.5735, 0.6695, 0.7406, 0.7180],
    [0.7359, 0.7372, 0.7383, 0.7997, 0.8184]
    ]
data = np.transpose(data)
X = np.arange(5) # the x locations for the groups
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, data[0], color="lightcoral", width=0.15)
ax.bar(X + 0.15, data[1], color="wheat", width=0.15)
ax.bar(X + 0.30, data[2], color="sienna", width=0.15)
ax.bar(X + 0.45, data[3], color="darkorange", width=0.15)
ax.bar(X + 0.60, data[4], color="gold", width=0.15)



ax.set_xticks(X+0.225, ("10", "20", "50", "75", "100"))
ax.legend(labels=["Baseline", "smallBERT", "smallBERT-F", "shotBERT", "shotBERT-F"])
plt.title("Macro few-shot results summary")

plt.show()
```

## Barplot simplified differences

### Macro

```python
import numpy as np
import matplotlib.pyplot as plt

baseline = [0.1365, 0.5303, 0.7359]

smallBERT = [0.4280, 0.6575, 0.7372]
smallBERTF = [0.4082, 0.7182, 0.7383]
shotBERT = [0.1000, 0.6920, 0.7997]
shotBERTF = [0.1079, 0.6421, 0.8184]

diff_smallBERT = np.array(smallBERT) - np.array(baseline) # absolute difference between baseline and smallBERT
diff_smallBERTF = np.array(smallBERTF) - np.array(baseline) # absolute difference between baseline and smallBERTF
diff_shotBERT = np.array(shotBERT) - np.array(baseline) # absolute difference between baseline and shotlBERT
diff_shotBERTF = np.array(shotBERTF) - np.array(baseline) # absolute difference between baseline and shotlBERTF



X = np.arange(3) # the x locations for the groups
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, diff_smallBERT, width=0.15, label='smallBERT')
ax.bar(X + 0.15, diff_smallBERTF, width=0.15, label='smallBERT+F')
ax.bar(X + 0.30, diff_shotBERT, width=0.15, label='shotBERT+C')
ax.bar(X + 0.45, diff_shotBERTF, width=0.15, label='shotBERT+CF')

ax.set_xticks(X+0.225, ("20", "50", "100"))

plt.xlabel('Number of shots')
plt.ylabel('F1-score difference')
plt.axhline(y=0, color='black', linestyle='--')  # horizontal line at y=0

plt.title("Macro few-shot results summary")
plt.legend()
plt.show()
```

### Weighted

```python
import numpy as np
import matplotlib.pyplot as plt

baseline = [0.1365, 0.5303, 0.7359]

smallBERT = [0.3011, 0.7218, 0.8480]
smallBERTF = [0.3050, 0.8192, 0.7546]
shotBERT = [0.0935, 0.8277, 0.8810]
shotBERTF = [0.1035, 0.7256, 0.8931]

diff_smallBERT = np.array(smallBERT) - np.array(baseline) # absolute difference between baseline and smallBERT
diff_smallBERTF = np.array(smallBERTF) - np.array(baseline) # absolute difference between baseline and smallBERTF
diff_shotBERT = np.array(shotBERT) - np.array(baseline) # absolute difference between baseline and shotlBERT
diff_shotBERTF = np.array(shotBERTF) - np.array(baseline) # absolute difference between baseline and shotlBERTF



X = np.arange(3) # the x locations for the groups
fig = plt.figure()
ax = fig.add_axes([0, 0, 1, 1])
ax.bar(X + 0.00, diff_smallBERT, width=0.15, label='smallBERT')
ax.bar(X + 0.15, diff_smallBERTF, width=0.15, label='smallBERT+F')
ax.bar(X + 0.30, diff_shotBERT, width=0.15, label='shotBERT+C')
ax.bar(X + 0.45, diff_shotBERTF, width=0.15, label='shotBERT+CF')

ax.set_xticks(X+0.225, ("20", "50", "100"))

plt.xlabel('Number of shots')
plt.ylabel('F1-score difference')
plt.axhline(y=0, color='black', linestyle='--')  # horizontal line at y=0

plt.title("Weighted few-shot results summary")
plt.legend()
plt.show()
```
